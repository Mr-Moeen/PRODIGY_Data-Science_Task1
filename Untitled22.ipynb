```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_84031.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
# Count NAN values
df.isna().sum()
```

```python
# fill missing values
df.ffill(inplace=True)

# Drop 'Unnamed: 5'
df.drop(columns=['Unnamed: 5'], inplace=True)
df
```

```python
df.isna().sum()
```

```python
# Catagorical chart
df.Region.value_counts().plot(kind='bar',
                             color='Blue',
                             width=0.9)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')
plt.show();
```

```python
# To plot the histogram for 'IncomeGroup'
plt.hist(df['IncomeGroup'], bins=10, color='yellow', edgecolor='black')

plt.xlabel('Income Group')
plt.ylabel('Count')
plt.title('Histogram of IncomeGroup')
plt.show();
```

```python
# create a continuous line chart showing the distribution of values in the 'IncomeGroup' column.

df['IncomeGroup'].value_counts().sort_index().plot(kind='line', marker='o', color='skyblue')

plt.xlabel('IncomeGroup')
plt.ylabel('Count')
plt.title('Distribution of IncomeGroup')

plt.show()
```
